```python
import pandas as pd

df1 = pd.DataFrame({
    'Student': ['Ali', 'Sara', 'Usman'],
    'Math': [85, 90, 78]
})

df2 = pd.DataFrame({
    'Student': ['Ali', 'Sara', 'Zoya'],
    'English': [80, 85, 90]
})

merged_df = pd.merge(df1, df2, on='Student', how='inner')
print(merged_df)
```
