```python
import matplotlib.pyplot as plt
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Generate random data
x = np.random.rand(1000)
y = np.random.rand(1000)

# Plot
plt.plot(x, y, color='red', linestyle='--', marker='o')  # smaller markers for clarity
plt.title('Sample plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```
