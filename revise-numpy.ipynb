```python
import matplotlib.pyplot as plt
import numpy as np 

# 2D grayscale image
image = np.array([
    [255, 200, 100, 200, 180],
    [140, 200, 120, 250, 240],
    [200, 120, 200, 250, 240]
])

plt.imshow(image, cmap='gray')  # cmap='gray' for grayscale
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

# Shape must be (height, width, 3)
image_rgb = np.array([
    [[255,0,0], [0,255,0], [0,0,255], [255,255,0], [0,255,255]],
    [[255,0,255], [128,128,0], [0,128,128], [128,0,128], [200,100,50]],
    [[50,200,100], [100,50,200], [200,100,150], [50,150,200], [250,250,250]]
], dtype=np.uint8)

plt.imshow(image_rgb)
plt.show()
```
